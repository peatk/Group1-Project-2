```python
import pandas as pd
from pathlib import Path
import sqlite3
import os
```

```python
path = Path("data/PEN America's Index of School Book Bans (July 1, 2021 - June 30, 2022).xlsx")
author_title_df = pd.read_excel(path, "Sorted by Author & Title")


author_title_df.head()
```

```python
author_title_df.dtypes
```

```python
author_title_df = author_title_df.fillna('')
author_title_df = author_title_df.loc[:, ['Author', 'Title', 'Type of Ban', 'State', 'District', 
                                          'Date of Challenge/Removal', 'Origin of Challenge', 'Latitude', 'Longitude' ]]

author_title_df= author_title_df.rename(columns={'Type of Ban':'Type_of_Ban',
                               'Date of Challenge/Removal': 'Date_of_Challenge_or_Removal',
                               'Origin of Challenge':'Origin_of_Challenge'})
```

```python
author_title_df.head()
```

```python
author_title_df.to_json('data/author_title.json', orient='records', force_ascii=False)
```

```python
state_district_df = pd.read_excel(path, "Sorted by State & District")
state_district_df = state_district_df.fillna('')
state_district_df.head()
state_district_df.dtypes
state_district_df = state_district_df.loc[:, ['State', 'District', 'Origin of Challenge','Date of Challenge/Removal',
                                              'Type of Ban', 'Author', 'Title', 
                                           'Latitude', 'Longitude' ]]

state_district_df= state_district_df.rename(columns={'Type of Ban':'Type_of_Ban',
                               'Date of Challenge/Removal': 'Date_of_Challenge_or_Removal',
                               'Origin of Challenge':'Origin_of_Challenge'})
state_district_df.head()
```

```python
state_district_df.to_json('data/state_district.json', orient='records',force_ascii=False)
```
