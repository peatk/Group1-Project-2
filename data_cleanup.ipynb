```python
state_district_df = pd.read_excel(path, "Sorted by State & District")
state_district_df = state_district_df.fillna('')
state_district_df.head()
state_district_df.dtypes
state_district_df = state_district_df.loc[:, ['State', 'District', 'Origin of Challenge','Date of Challenge/Removal',
                                              'Type of Ban', 'Author', 'Title', 
                                           'Latitude', 'Longitude' ]]

state_district_df= state_district_df.rename(columns={'Type of Ban':'Type_of_Ban',
                               'Date of Challenge/Removal': 'Date_of_Challenge_or_Removal',
                               'Origin of Challenge':'Origin_of_Challenge'})
state_district_df.head()
```

```python
state_district_df.to_json('data/state_district.json', orient='records',force_ascii=False)
```

```python
state_district_group_df = state_district_df.groupby('District').first().reset_index()
state_district_group_df
```

```python
import requests
import numpy as np

# Your Geoapify API key
api_key = "bc76e695f9e54cebae157ec8551f0c47"


dis_lat = []
dis_lng = []
# Assuming 'state_district_df' is your DataFrame and it has columns 'District' and 'State'
for index, row in state_district_group_df.iterrows():
    district = row['District']
    state = row['State']  # Ensure this column exists in your DataFrame
    # Constructing the request URL
    url = f"https://api.geoapify.com/v1/geocode/search?text={district}, United States&apiKey={api_key}"
    
    # Sending the request and storing the response object
    response = requests.get(url)
    
    # Checking if the request was successful
    if response.status_code == 200:
        data = response.json()
        matched = False  # Initialize flag to check if we found a matching state
        
        # Iterate through each feature to find a matching state
        for feature in data['features']:
            # Some responses might not have 'state' in 'properties', so use get() with a default value
            feature_state = feature['properties'].get('state', '').lower()
            if feature_state == state.lower():
                lat = feature['geometry']['coordinates'][1]
                lng = feature['geometry']['coordinates'][0]
                print(f"{district}: Latitude {lat}, Longitude {lng}")
                dis_lat.append(lat)
                dis_lng.append(lng)
                matched = True
                break  
        
        if not matched:
            lat = "N/A"
            lng = "N/A"
            print(f"{district}: Latitude {lat}, Longitude {lng}")
            dis_lat.append(lat)
            dis_lng.append(lng)
            
    else:
        print(f"Failed to fetch data for {district}, Status code: {response.status_code}")

state_district_group_df["Dis_lat"] = dis_lat
state_district_group_df["Dis_lng"] = dis_lng
state_district_group_df.head()

```

```python
state_district_group_df = state_district_group_df[["District", "State", "Dis_lat", "Dis_lng"]]
state_district_group_df
```

```python
state_district_group_df.to_csv('distric_lat_lng.csv', index=False)
```
