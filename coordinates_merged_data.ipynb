```python
import pandas as pd

# Load the CSV files into pandas DataFrames
lat_lng_df = pd.read_csv('data/statedistrict_lat_lng.csv')
state_district_df = pd.read_csv('data/state_district.csv', index_col=0)

lat_lng_df.head()
```

```python
state_district_df.head()
```

```python
# Merging the DataFrames on the "District" column
merged_df = pd.merge(state_district_df, lat_lng_df[['District', 'Dis_Latitude', 'Dis_Longitude']], on="District", how="left")

# Display the first few rows of the merged DataFrame to verify the merge
merged_df.head()
```

```python
merged_df.to_json('data/coordinates.json', orient='records',force_ascii=False)
```
