Bar graph could showcase the amount of times a book has been banned in the US. Limit it to the top 10 books, or top 10 authors (book count per author). 


```python
import pandas as pd
from pathlib import Path

import os
```

```python
df = pd.read_json("data/author_title.json", orient='records')
df.head()
```

```python
gdf = df.groupby(["Title"])
title_df = gdf.count().sort_values(["State"], ascending=False)["State"]
title_df = title_df.head(10)
title_df = title_df.iloc[::-1]
title_df.plot(kind='barh',title="Top books banned")
```

```python
gdf = df.groupby(["Author"])
author_df = gdf.count().sort_values(["State"], ascending=False)["State"]
author_df = author_df.head(10)
author_df = author_df.iloc[::-1]
author_df.plot(kind='barh', title="Top author banned")
```
